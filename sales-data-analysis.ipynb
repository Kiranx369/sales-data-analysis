```python
import pandas as pd

data = {
    "city": ["Pune","Mumbai","Delhi","Pune","Mumbai","Delhi","Pune","Delhi","Mumbai","Pune","Mumbai","Delhi","Pune","Mumbai","Delhi"],
    "sales": [200,300,250,400,150,500,350,450,380,220,330,270,410,180,520],
    "month": ["Jan","Jan","Jan","Feb","Feb","Feb","Mar","Mar","Mar","Apr","Apr","Apr","May","May","May"],
    "product": ["A","B","A","C","A","B","C","A","C","B","C","A","B","A","C"],
    "category": ["Electronics","Furniture","Electronics","Clothing","Furniture","Electronics","Clothing","Electronics","Clothing","Furniture","Clothing","Electronics","Furniture","Electronics","Clothing"]
}

df = pd.DataFrame(data)

df
```

```python
df.groupby("city")["sales"].sum()
```

```python
df.groupby("month")["sales"].sum()
```

```python
df.groupby("category")["sales"].sum()
```

```python
df.pivot_table(values="sales",index="city",columns="month",aggfunc="sum")
```

```python
import matplotlib.pyplot as plt

df.groupby("city")["sales"].sum().plot(kind="bar")
plt.title("Total Sales by City")
plt.show()
```
